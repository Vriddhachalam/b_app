```python
import json
import os
import time

INPUT_FILE = "assets/songs.json"
OUTPUT_FILE = "assets/sairhythms_full_songs_temp.json"
FLUSH_EVERY = 10

# ---- load song index ----
with open(INPUT_FILE, "r", encoding="utf-8") as f:
    raw = json.load(f)

items = json.loads(raw["data"])

# ---- load existing output if present ----
if os.path.exists(OUTPUT_FILE):
    with open(OUTPUT_FILE, "r", encoding="utf-8") as f:
        all_songs = json.load(f)
    scraped_slugs = {s.get("slug") for s in all_songs}
else:
    all_songs = []
    scraped_slugs = set()

print(f"Already have {len(scraped_slugs)} songs")

# ---- scrape loop ----
count = 0

for i, item in enumerate(items, 1):
    url = item.get("url")
    if not url or not url.startswith("/song/"):
        continue

    slug = url.replace("/song/", "").strip()

    # skip already scraped
    if slug in scraped_slugs:
        continue

    try:
        detailed = scrape_song(slug)

        merged = {**item, **detailed}

        # normalize text junk
        for k, v in merged.items():
            if isinstance(v, str):
                merged[k] = v.replace("\xa0", " ").strip()

        all_songs.append(merged)
        scraped_slugs.add(slug)
        count += 1

        print(f"[{i}/{len(items)}] scraped {slug}")

        # ---- flush every N songs ----
        if count % FLUSH_EVERY == 0:
            with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
                json.dump(all_songs, f, ensure_ascii=False, indent=2)
            print(f"ğŸ’¾ saved {len(all_songs)} songs to disk")

        time.sleep(0.5)  # polite scraping

    except Exception as e:
        print(f"[FAIL] {slug}: {e}")

# ---- final write ----
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(all_songs, f, ensure_ascii=False, indent=2)

print(f"âœ… done. total songs saved: {len(all_songs)}")
```

```python
import requests
from bs4 import BeautifulSoup

def clean_text(el):
    return el.get_text(" ", strip=True)

def scrape_song(slug):
    url = f"https://sairhythms.sathyasai.org/song/{slug}"
    r = requests.get(url, timeout=10)
    r.raise_for_status()

    soup = BeautifulSoup(r.text, "html.parser")

    data = {
        "slug": slug,
        "url": url
    }

    container = soup.select_one(".devotional-song-meta-data")
    if not container:
        return data

    # ---- generic metadata rows ----
    for row in container.select(".devotional-song-meta-data-grid"):
        cols = row.find_all("div", recursive=False)
        if len(cols) < 2:
            continue

        key = clean_text(cols[0])
        val = clean_text(cols[1])

        if key and val:
            data[key] = val

    # ---- audio ----
    audio = soup.select_one("audio source")
    if audio:
        data["audio"] = audio["src"]

    # ---- youtube ----
    yt = soup.select_one(".devotional-song-video-link a")
    if yt:
        data["youtube"] = yt["href"]

    # ---- raga notes (expanded panel) ----
    raga_panel = soup.select_one(".raga-notes-panel")
    if raga_panel:
        data["raga_notes"] = clean_text(raga_panel)

    # ---- glossary ----
    glossary = {}
    glossary_items = soup.select(".devotional-song-glossary-term")
    for term_div in glossary_items:
        term = term_div.get_text(strip=True).rstrip(":")
        desc_div = term_div.find_next_sibling(
            "div", class_="devotional-song-glossary-term-description"
        )
        if term and desc_div:
            glossary[term] = clean_text(desc_div)

    data["glossary"] = glossary

    return data
```

```python
import json

PREFIX = "https://sairhythms.org"
PREFIX1 = "soundcloud"
PREFIX2 = "raagabox"


with open("assets/songs.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Your data["data"] is a JSON string, not a list
items = json.loads(data["data"])

invalid_songs = []

for song in items:
    audio_links = song.get("audio_link", [])

    for link in audio_links:
        if not (link.startswith(PREFIX) or PREFIX1 in link or 'youxtu' in link or 'sundaram' in link):
            invalid_songs.append({
                "song_id": song.get("song_id"),
                "title": song.get("title"),
                "audio_link": link
            })
            break  # one bad link is enough to flag the song

print(f"Total songs: {len(items)}")
print(f"Songs NOT starting with '{PREFIX}': {len(invalid_songs)}")

# Show first 10 problematic songs
for s in invalid_songs:
    print(s)
```
